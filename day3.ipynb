#So, the mul(a,b) function is simple — it just multiplies two numbers.

Now, df.apply() wants a function that takes each row of the dataframe and does something with it.

But mul alone only knows how to multiply two numbers, not how to handle a whole row.

That’s why you do:

python
Copy
Edit
df.apply(lambda row: mul(row["age"], row["newage"]), axis=1)
Here:

lambda row: ... is an anonymous function — a tiny function that takes one argument, row.

That row represents each row of the dataframe as apply goes through it.

Inside the lambda, you call mul with row["age"] and row["newage"].

So yeah, lambda is a function, just an unnamed one, letting you grab the row and then use mul for the multiplication.

If you tried just df.apply(mul), it wouldn’t work because mul expects two separate arguments, but apply only gives each row as a single argument.

So the lambda wraps around to take that row, pick the two needed values, and then feed them into mul.
